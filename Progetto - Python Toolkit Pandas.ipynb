### PROGETTO DISASTRI AEREI 
Importazione del dataset utilizzando Pandas

```python
import pandas as pd 
dates = pd.read_csv("aviation-accidents.csv")
```

```python
print(dates.isnull()) #Controllo di eventuali celle vuote che inficierebbero sullo studio in oggetto
```

```python
dates_clean = dates.dropna() #eliminazione delle righe contenenti celle vuote
dates_clean
```

```python
print(dates_clean.isnull()) 
```

```python
dates_clean.isnull().sum() #controllo di celle vuote
```

```python
dates_clean['date'] = pd.to_datetime(dates_clean['date'], errors = 'coerce')
dates_clean_data = dates_clean.dropna(subset=['date'])
dates_clean_data #cambio tipo di dato per la colonna "date", eliminando le date errate e "unknown"
```

```python
print(dates_clean_data.isnull())
```

```python
dates_clean_data.isnull().sum() #ricontrollo ddel valori nulli presenti nel dataset
```

```python
dates_clean_data['fatalities'].unique() #la colonna fatalities presenza delle celle contenti il simbolo '+'
```

```python
def convert_fatalities(value):
    # divide la stringa basata su "+", poi converte i pezzi in interi e li somma
    # questo funziona sia per stringhe singole tipo '12' sia per somme tipo '3+4'
    return sum(int(part) for part in value.split("+"))

# Applica la funzione a ogni valore nella colonna 'fatalities'
dates_clean_data['fatalities'] = dates_clean_data['fatalities'].apply(convert_fatalities)
```

Andiamo a creare un df da poter utilizzare nei grafici inserendo la colonna del totale delle vittime per operatore e la frequenza di quante volte l'operatore si presente all'interno del df principale

```python
sum_fatalities_per_operator = dates_clean_data.groupby('operator')['fatalities'].sum().reset_index()
sum_fatalities_per_operator.columns = ['Operator', 'Total Fatalities']
operator_frequency = dates_clean_data['operator'].value_counts().reset_index()
operator_frequency.columns = ['Operator', 'Frequency']
sum_fatalities_with_frequency = pd.merge(sum_fatalities_per_operator, operator_frequency, on='Operator')
# Visualizziamo il nuovo DataFrame
print(sum_fatalities_with_frequency)
```

```python
sorted_df = sum_fatalities_with_frequency.sort_values('Total Fatalities', ascending=False)
# Visualizza il DataFrame ordinato
print(sorted_df)
```

```python
# Ottieni i primi 15 record per Total Fatalities
top_15_fatalities = sorted_df.head(15)
# Visualizza i primi 15 record
print(top_15_fatalities)
```

```python
plt.figure(figsize=(15,8)) #visualizzazione grafico
bars = plt.barh(top_15_fatalities['Operator'], top_15_fatalities['Total Fatalities'], color=personal_colors)
plt.xlabel('Total Fatalities')
plt.ylabel('Operator')
plt.title('Top 15 Operators by Total Fatalities')
plt.gca().invert_yaxis() 
for bar in bars:
    width = bar.get_width()
    label_x_pos = width + bar.get_width() * 0.01  # Sposta leggermente il testo a destra per maggiore visibilità
    plt.text(label_x_pos, bar.get_y() + bar.get_height()/2, s=f'{width}', va='center')
for i, (val, op) in enumerate(zip(top_15_fatalities['Total Fatalities'], top_15_fatalities['Operator'])):
    plt.barh(op, val, color=personal_colors[i], label=op)
plt.legend(loc='lower right', title="Operators") #legenda
plt.show()
```

Dal grafico sopra visualizzato possiamo vedere che l'operatore con il maggior numero di vittime è: USAF per un totale di 5801, successivamente si ha USAAF con 4509 ed american airlines con 3164.

```python
df_accident = dates_clean_data.copy() #riprendiamo il df iniziale
df_accident
```

```python
import geopandas as gpd #importazione della libreria
gdf_world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres')) #geodata con le geometrie dei paesi
df_aggregated = df_accident.groupby('country').size().reset_index(name='incident_count') #aggregazione
gdf_merged = gdf_world.merge(df_aggregated, left_on='name', right_on='country') #Unisci il GeoDataFrame dei paesi con i tuoi dati
fig, ax = plt.subplots(1, 1, figsize=(25, 10)) #Visualizza i dati sulla mappa
gdf_world.boundary.plot(ax=ax, linewidth=1)
gdf_merged.plot(column='incident_count', ax=ax, legend=True,
                legend_kwds={'label': "Accidents for each country"},
                cmap='OrRd') 

plt.show()
```

Dalla mappa possiamo notare la rappresentazione degli stati elencati nel df originale. per i colori rosso tendente allo scuro abbiamo una maggiore presenza di incidenti come ad esempio la Russia per un totale di 1.177 incidenti. Ma non tutti gli stati sono matchati

```python
#replace nel df degli stati che hanno nomi differenti rispetto al file gdf_world con le geometrie
df_accident['country'] = df_accident['country'].replace('USA', 'United States of America') 
df_accident['country'] = df_accident['country'].replace('U.K.', 'United Kingdom')
```

```python
import geopandas as gpd #importazione della libreria
gdf_world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres')) #geodata con le geometrie dei paesi
df_aggregated = df_accident.groupby('country').size().reset_index(name='incident_count') #aggregazione
gdf_merged = gdf_world.merge(df_aggregated, left_on='name', right_on='country') #Unisci il GeoDataFrame dei paesi con i tuoi dati
fig, ax = plt.subplots(1, 1, figsize=(25, 10)) #Visualizza i dati sulla mappa
gdf_world.boundary.plot(ax=ax, linewidth=1)
plot = gdf_merged.plot(column='incident_count', ax=ax, legend=True,
                cmap='RdBu') 
legend = ax.legend(title="Accidents for each country", loc='upper center', bbox_to_anchor=(0.5, 1.05),
                   ncol=3, shadow=True, fancybox=True, borderpad=1)
legend.get_title().set_fontsize('15') 
legend.get_frame().set_edgecolor('blue')
legend.get_frame().set_linewidth(2)
plt.xlabel("Longitudine", fontsize=25)
plt.ylabel("Latitudine",fontsize=25)
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.savefig('nome_file.png', dpi=300) #salvare la figura in png
plt.show()
```

Dal grafico possiamo notare visivamente come gli incidenti più frequenti siano stati registrati nel stati uniti come nel primo grafico mostrato dalla nostra analisi, seguito subito dalla Russia con un colore più chiaro ed al terzo posta il canada.
